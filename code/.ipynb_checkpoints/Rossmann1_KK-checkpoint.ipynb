```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

pd.options.display.max_rows=None
pd.options.display.max_columns=None

import warnings
warnings.filterwarnings('ignore')
```

```python
class color:
   PURPLE = '\033[95m'
   CYAN = '\033[96m'
   DARKCYAN = '\033[36m'
   BLUE = '\033[94m'
   GREEN = '\033[92m'
   YELLOW = '\033[93m'
   RED = '\033[91m'
   BOLD = '\033[1m'
   UNDERLINE = '\033[4m'
   END = '\033[0m'
```

```python
sales=pd.read_csv("../input/train.csv")

print(sales.shape)

print(color.BOLD+"\n No. of null data in sales \n \n"+color.END+" {}".format(sales.isnull().sum()))

print(color.BOLD+"\nStores overview"+color.END)

sales.head()
```

```python
store=pd.read_csv("../input/store.csv")

print(store.shape)

print(color.BOLD+"\n \n No. of null data in stores \n \n "+color.END+" {}".format(store.isnull().sum()))

print(color.BOLD+"\nStores overview"+color.END)

store.head()
```

```python
print(color.BOLD+"Observing records for a single Store."+color.END)
a=sales[sales['Store']==1]
a[a["Open"]==1]
```

```python
print("Unique stores record in sales: {}".format(len(sales["Store"].unique())))
```

```python
print("{} - Decimal, hence records are not equally distributed among all stores".format(sales.shape[0]/len(sales["Store"].unique())))
```

```python
fig,ax = plt.subplots(figsize=(15,5))
fig.suptitle('# Sales records for each Store', fontsize=30, fontweight='bold')
ax.set_xlabel('Stores')
ax.set_ylabel('Records')
plt.bar(sales["Store"].unique(),sales.groupby(["Store"])["Store"].count())
```

**The above graph shows that, not all stores have data for 30 months**

```python
print("The different records(1M+) are spread across how how many days?")
print(len(sales["Date"].unique()))
```

```python
MasterData=pd.merge(sales, store, how='left', on="Store")
```

```python
salesrecords=MasterData.groupby(["Store"]).count()["DayOfWeek"].unique()
salesrecords=list((map(str,salesrecords)))
salesrecords
```

```python
Storecount=[(MasterData.groupby(["Store"]).count()["DayOfWeek"]==942).sum(),(MasterData.groupby(["Store"]).count()["DayOfWeek"]==758).sum(),(MasterData.groupby(["Store"]).count()["DayOfWeek"]==941).sum()]
```

```python
fig,ax = plt.subplots(figsize=(5,4))
fig.suptitle('# of Stores with # of Records', fontsize=20, fontweight='bold')
ax.set_xlabel('Sales Records')
ax.set_ylabel('Stores count')
plt.bar(salesrecords,Storecount)
```

```python
MasterData["Year"]=MasterData.Date.apply(lambda x: x.split('-')[0])
MasterData["Month"]=MasterData.Date.apply(lambda x: x.split('-')[1])
```

```python
fig,ax = plt.subplots(figsize=(15,5))
fig.suptitle('Sales of each store for given years', fontsize=20, fontweight='bold')
ax.set_xlabel('Store and Year')
ax.set_ylabel('Sales')
MasterData.groupby(["Store","Year"])["Sales"].sum().plot()
```

```python
MasterData.head()
```

```python
fig,ax = plt.subplots(figsize=(15,5))
fig.suptitle('Sales per year', fontsize=20, fontweight='bold')
ax.set_xlabel('Year')
ax.set_ylabel('Sales')
MasterData.groupby(["Year"])["Sales"].sum().plot.bar()
```

**The trend shows slight decline in sales for 2014 and a heavy decline for 2015**

Lets look at sales month wise

```python
fig,ax = plt.subplots(figsize=(15,5))
fig.suptitle('Sales each Month for given years', fontsize=20, fontweight='bold')
ax.set_xlabel('Month and Year')
ax.set_ylabel('Sales')
MasterData.groupby(["Year","Month"])["Sales"].sum().plot.bar()
```

Thats because for 2015 we have only 7 month's data

```python
fig,ax = plt.subplots(figsize=(15,5))
fig.suptitle('Sales each Month for given years', fontsize=20, fontweight='bold')
ax.set_xlabel('Month and Year')
ax.set_ylabel('Sales')
MasterData.groupby(["Month","Year"])["Sales"].sum().plot.bar()
```
