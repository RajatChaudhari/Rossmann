### Analysis on Stores, Sales and Date

```python
print("{} - Decimal, hence records are not equally distributed among all stores".format(sales.shape[0]/len(sales["Store"].unique())))
```

```python
fig,ax = plt.subplots(figsize=(15,5))
fig.suptitle('# Sales records for each Store', fontsize=30, fontweight='bold')
ax.set_xlabel('Stores')
ax.set_ylabel('Records')
plt.bar(sales["Store"].unique(),sales.groupby(["Store"])["Store"].count())
```

```python
print("The different records(1M+) are spread across how how many days?")
print(len(sales["Date"].unique()))
```

```python
MasterData=pd.merge(sales, store, how='left', on="Store")
```

```python
salesrecords=MasterData.groupby(["Store"]).count()["DayOfWeek"].unique()
salesrecords=list((map(str,salesrecords)))
salesrecords
```

```python
Storecount=[(MasterData.groupby(["Store"]).count()["DayOfWeek"]==942).sum(),(MasterData.groupby(["Store"]).count()["DayOfWeek"]==758).sum(),(MasterData.groupby(["Store"]).count()["DayOfWeek"]==941).sum()]
```

###### this shows that all stores are not open on all given days, some stores are open 942 days, some 941 and some only for 758 days

```python
fig,ax = plt.subplots(figsize=(5,4))
fig.suptitle('# of Stores with # of Records', fontsize=20, fontweight='bold')
ax.set_xlabel('Sales Records')
ax.set_ylabel('Stores count')
plt.bar(salesrecords,Storecount)
```
